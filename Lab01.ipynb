```python
# Histogram of Agility of Players by Age
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
agility = df.Agility
ages = df.Age
ax.bar(ages, agility)
ax.set_title('Agility of Players by Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Agility')
plt.show()
```

```python
# Players from 16 to 34 are very agile compared to players 35 and up.
```

# Correlation

```python
corr = df.corr()
corr
```

# Positive Correlations

# Reaction and Age are correlated by about 0.45, this makes sense because reaction time probably isn't the best when a player is very young or on the older side of the spectrum.
# Standing tackle and sliding tackle are correlated by about 0.97 because they're both defensive moves that are used often in soccer.
# Gk Handling and GK Diving are correlated by about 0.96 because the goalie has to dive a lot to get the ball.


# Negative Correlations

# Jersey Number and Overall are negatively correlated because they have a total of -0.25 since they have almost nothing to do with each other.
# Dribbling and GK Handling are negatively correlated with a score of about about -0.75 because goal keepers mainly don't dribble the ball so it's not very related.
# Finishing and Jersey Number are negatively correlated with a score of -0.006639 which means the data was not related but not super unrelated otherwise the number would be close to -1.

# Columns I would use would be ones that are logically related in terms of position and defensive/offensive moves. For example I would look at GK Handling, Diving, Kicking, Positioning and Reflexes because these categories specifically focus on Goal Keepers.
